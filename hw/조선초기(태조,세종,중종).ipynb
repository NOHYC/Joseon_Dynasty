### 언급이 많이 된 단어 TOP15

```python
counts = pd.DataFrame(word_count)
counts.columns= ['단어','빈도수']
king_11 = counts.sort_values('빈도수',ascending=False)[:15]
```

```python
plt.figure(figsize=(15,10))
plt.xticks(rotation = 45 )
sns.barplot(x=king_11.단어, y=king_11.빈도수)
```
